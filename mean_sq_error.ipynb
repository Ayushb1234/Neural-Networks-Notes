```python
def mse(actual, predicted):
    
    n = len(actual)

    error = 0

    for a, p in zip(actual, predicted):
        error += (a - p) ** 2

    return error / n


actual = [1, 2, 3]
predicted = [12, 18, 33]

print(mse(actual, predicted))
```

```python
# gradient descent
# ------------------------


def update_wt(weight, gradient, lr):
    return weight - lr*gradient

print(update_wt(8,3,0.1))
```
