```python
# learning rate
# --------------

def compute(weight,gradient,lr):
    return weight - lr*gradient

print(compute(10, 2, 0.1))
print(compute(10, 2, 1))
print(compute(10, 2, 10))
    
```

```python
# derivates
# ---------------

def derivates(x):
    return 2*x

print(derivates(3))
print(derivates(10))
```
